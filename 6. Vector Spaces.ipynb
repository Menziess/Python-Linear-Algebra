## 1. Solutions To Linear Systems

We have 3 points in 2d space, and we would like to determine a second degree polynomial that passes through all of these points, having the form of:
$p(\chi) = \gamma_0 + \gamma_1\chi + \gamma_2\chi^2$

```python
points = np.array([(-2, -1), (0, 2), (2, 3)]
```

```python
plt.scatter(*zip(*points), color='y')
plt.xlim(-2.5, 2.5)
plt.ylim(-1.5, 3.5)
plt.grid()
plt.show()
```

### 1.1 A Solution

We know that:
    
- $p(-2) = -1$
- $p(0) = 2$
- $p(2) = 3$

So we can set up these equations:

$
\begin{array}{}
p(-2) & = & \gamma_0 & + & \gamma_1(-2) & + & \gamma_2(-2)^2 & = & -1 \\
p(0) & = & \gamma_0 & + & \gamma_1(0) & + & \gamma_2(0)^2 & = & 2 \\
p(2) & = & \gamma_0 & + & \gamma_1(2) & + & \gamma_2(2)^2 & = & 3 \\
\end{array}
$

This linear system can be defined as:

$
\begin{bmatrix}
1 & -2 & 4 \\
1 & 0 & 0 \\
1 & 2 & 4 \\
\end{bmatrix}
$
$
\begin{bmatrix}
\gamma_0 \\
\gamma_1 \\
\gamma_2
\end{bmatrix}
$
=
$
\begin{bmatrix}
-1 \\
2 \\
3
\end{bmatrix}
$

Solving this system yields:

$
\begin{bmatrix}
\gamma_0 \\
\gamma_1 \\
\gamma_2
\end{bmatrix}
$
=
$
\begin{bmatrix}
2 \\
1 \\
-\frac{1}{4}
\end{bmatrix}
$

So that:

$p(\chi) = 2 + \chi - \frac{1}{4}\chi^2$

You can capture the polynomials at the values $-2, 0,$ and $2$:

```python
def p(x):
    return 2 + x - 0.25 * x**2

def gamma0(x):
    return 1
def gamma1(x):
    return x
def gamma2(x):
    return x**2

x = np.linspace(points[0][0], points[-1][0])
y = list(map(p, x))
g0 = list(map(gamma0, x))
g1 = list(map(gamma1, x))
g2 = list(map(gamma2, x))

plt.scatter(*zip(*points), color='y')
plt.plot(x, y, color='y')
plt.plot(x, g0, color='g')
plt.plot(x, g1, color='b')
plt.plot(x, g2, color='r')

plt.xlim(-2.5, 2.5)
plt.ylim(-1.5, 3.5)
plt.grid()
plt.show()
```

The question becomes, when is such a system solvable?

### 1.2 No Solution

What if we had an extra point?

```python
points = np.array([(-2, -1), (0, 2), (2, 3), (4, 9)])
```

We would have known that:
    
- $p(-2) = -1$
- $p(0) = 2$
- $p(2) = 3$
- $p(4) = 9$

Meaning that our polynomial would have to go through all of these points. And we would end up with four equations and three unknowns:

$
\begin{bmatrix}
1 & -2 & 4 \\
1 & 0 & 0 \\
1 & 2 & 4 \\
1 & 4 & 16 \\
\end{bmatrix}
$
$
\begin{bmatrix}
\gamma_0 \\
\gamma_1 \\
\gamma_2
\end{bmatrix}
$
=
$
\begin{bmatrix}
-1 \\
2 \\
3 \\
9 \\
\end{bmatrix}
$

We have already found a solution for the unknowns, so we can plug it in and see if 9 pops out:

$
\begin{bmatrix}
1 & 4 & 16 \\
\end{bmatrix}
$
$
\begin{bmatrix}
2 \\
1 \\
-\frac{1}{4}
\end{bmatrix}
\neq
\begin{bmatrix}
9 \\
\end{bmatrix}
$

As we can see, it does not.
